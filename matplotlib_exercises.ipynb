```python
x = range(2,50)
y = np.sqrt(x)
plt.plot(x,y)
plt.show()

x2 = range(1,50)
y2 = [i ** 3 for i in x2]
plt.plot(x2, y2)
plt.show()

x3 = range(1, 50)
y3 = [2 ** n for n in x3]
plt.plot(x3, y3)
plt.show()

x4 = range(1,50)
y4 = [1 / (n+1) for n in x4]
plt.plot(x4, y4)
plt.show()
```

Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(10,6))

plt.subplot(221)
plt.plot(x,y)

plt.subplot(222)
plt.plot(x2,y2)

plt.subplot(223)
plt.plot(x3,y3)

plt.subplot(224)
plt.plot(x4,y4)

plt.show()
```
